```python
#Importing the dependencies
import pandas as pd
import numpy as np
from textblob import TextBlob
import re
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
df=pd.read_csv('Data.csv', encoding = "ISO-8859-1")
```

```python
df.head()
```

```python
df.shape
```

```python
# Combine the top news headlines
headlines=[]


for row in range(0,len(df.index)):
    headlines.append(' '.join(str(x) for x in df.iloc[row,2:27]))
```

```python
headlines[0]
```

```python
#clean the data

clean_headlines=[]

for i in range(0, len(headlines)):
    clean_headlines.append(re.sub("'",'',headlines[i]))
    clean_headlines[i]=re.sub("'",'',clean_headlines[i])
    clean_headlines[i]=re.sub(",",'',clean_headlines[i])
    
    clean_headlines[i]=re.sub(":",'',clean_headlines[i])
```

```python
clean_headlines[0]
```

```python
df['Combined_News']=clean_headlines
```

```python
df.head()
```

```python
# Create functions to get the subjectivity

def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

 #Create a function to get the polarity

def getPolarity(text):
    return TextBlob(text).sentiment.polarity 
```

```python
df['subjectivity']=df['Combined_News'].apply(getSubjectivity)
df['Polarity']=df['Combined_News'].apply(getPolarity)
```

```python
df.head(3)
```

```python
# Create a function to get the sentiment scores 
def getSIA(text):
    sia=SentimentIntensityAnalyzer()
    sentiment=sia.polarity_scores(text)
    return sentiment
```

```python
#Get the sentiment scores for each day
compound=[]
neg=[]
pos=[]
neu=[]
SIA=0

for i in range(0,len(df['Combined_News'])):
    SIA=getSIA(df['Combined_News'][i])
    compound.append(SIA['compound'])
    neg.append(SIA['neg'])
    neu.append(SIA['neu'])
    pos.append(SIA['pos'])
```

```python
# Store the sentiment score 

df['compound']=compound
df['Negative']=neg
df['Neutral']=neu
df['Positive']=pos
```

```python
df.head(3)
```

```python
keep_columns=['subjectivity','Polarity','compound','Negative','Neutral','Positive','Label']
df1=df[keep_columns]
df1
```

```python
# Creating the feature dataset
X=df1
X=df1.drop(columns='Label',axis=1)

# Create the target data set
Y=df1['Label']
```

```python
# splitting the data into the test and train set

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
#Create and train the models

model=LinearDiscriminantAnalysis().fit(X_train, Y_train)
```

```python
# Show the model predictions
predictions=model.predict(X_test)
predictions
```

```python
Y_test
```

```python
#Show the model metrics
print(classification_report(Y_test,predictions))
```
